# About Dataset
The datasets provided include the players data for the Career Mode FIFA 22

```python
#Loading the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv("players_22.csv")
```

```python
fifa.head()
```

```python
for col in fifa.columns:
  print(col)
```

```python
fifa.shape
```

```python
fifa['nationality_name'].value_counts()
```

```python
fifa['nationality_name'].value_counts()[0:10]
```

```python
fifa['nationality_name'].value_counts().keys()
```

```python
plt.figure(figsize=(16,5))
plt.bar(list(fifa['nationality_name'].value_counts()[0:10].keys()),
        list(fifa['nationality_name'].value_counts()[0:10]),color="g")
plt.show()
```

```python
players_salary = fifa[['short_name','wage_eur']]
```

```python
players_salary.head()
```

```python
players_salary=players_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
players_salary.head()
```

```python
plt.figure(figsize=(16,5))
plt.bar(list(players_salary['short_name'][0:10]),
        list(players_salary['wage_eur'][0:10]),color="g")
plt.show()
```
